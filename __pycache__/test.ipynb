```python
def reshape_input(x, nuts=4):
    '''
    reshape vectors to matrices 
    with the shape of (nuts, ...)
    '''
    if x.shape[0]%nuts == 0:
        x_new = x.reshape(nuts, -1)
    else:
        o_length = x.shape[0]
        n_dimension = int(o_length - (o_length%nuts))
        x_new = x.reshape(nuts, n_dimension)
    return x_new 
```
